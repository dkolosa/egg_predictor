Lets try to perform feature expration and predict the amount of eggs per day a chicken can have

```python
data_eggs = pd.concat([data['Age'], data['AmountOfFeed'], data['SunLightExposure']], axis=1)
data_labels = data['EggsPerDay']
```

```python
data_eggs.tail()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import linear_model
from sklearn.metrics import accuracy_score, r2_score, mean_squared_error
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_eggs, data_labels, test_size=0.33, random_state=42, shuffle=True)
```

```python
lasso = linear_model.Lasso(alpha=.5)
lasso.fit(x_train, y_train)
y_pred = lasso.predict(x_test)
print(f'r2: {r2_score(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
```

```python
ridge = linear_model.Ridge(alpha=.5)
ridge.fit(x_train, y_train)
y_pred = ridge.predict(x_test)
print(f'r2: {r2_score(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
neighbor = KNeighborsRegressor()
neighbor.fit(x_train, y_train)
y_pred = neighbor.predict(x_test)
print(f'r2: {r2_score(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
```
